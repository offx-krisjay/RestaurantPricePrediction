```python
'''removing non numeric characters from column to convert into integer value'''
df['cost']=df['cost'].astype(str)
df['cost']=df['cost'].apply(lambda x:x.replace(',',''))
df['cost']=df['cost'].astype(float)
```

```python
df.loc[df.rating==0,'rating']=np.NaN
df.isna().sum()
```

```python
x=df.drop(['cost'],axis=1)
y=df['cost']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
rt=RandomForestRegressor()
from sklearn.metrics import mean_absolute_percentage_error,r2_score,mean_squared_error
```

```python
rt.fit(xtrain,ytrain)
ypred=rt.predict(xtest)
print(mean_absolute_percentage_error(ytest,ypred))
print(r2_score(ytest,ypred))
```

```python
print(mean_squared_error(ytest,ypred)**0.5)
```
